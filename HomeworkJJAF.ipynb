1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r')
```

```python
archivo
```

```python
cabecera = archivo.readline()
print(cabecera)
nombres_variables = cabecera.strip().split('|')
print(nombres_variables)
cantidad_variables = len(nombres_variables)
print(cantidad_variables)
print("Hay", cantidad_variables, "variables")
```

```python
next(archivo)
tipos_de_datos = {}

for linea in archivo:
        valores = linea.strip().split('|')
        for i in range(len(nombres_variables)):
            nombre_variable = nombres_variables[i]
            valor = valores[i]
            if valor.isdigit():
                tipo = 'Integer'
            else:
                try:
                    float(valor)
                    tipo = 'Float'
                except ValueError:
                    tipo = 'String'
            if nombre_variable not in tipos_de_datos:
                tipos_de_datos[nombre_variable] = tipo
            else:
                tipo_actual = tipos_de_datos[nombre_variable]
                if tipo_actual != tipo:
                    tipos_de_datos[nombre_variable] = 'String'
print("Tipos de datos de cada variable:")
for variable, tipo in tipos_de_datos.items():
    print(variable + ": " + tipo)
```

```python
valores_faltantes_por_columna = {}
cabecera
columnas = cabecera.split('|')

for col in columnas:
    valores_faltantes_por_columna[col] = 0
archivo = open('Emisiones_CO2.csv', 'r')
next(archivo)
for linea in archivo:
        fila = linea.strip().split('|')
        for col_idx, valor in enumerate(fila):
            if valor == '':
                valores_faltantes_por_columna[columnas[col_idx]] += 1

print("Valores faltantes por columna:")
for columna, num_faltantes in valores_faltantes_por_columna.items():
    print(columna + ": " + str(num_faltantes))
```
